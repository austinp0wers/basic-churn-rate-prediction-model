```python
# !pip install tensorflow

# import datasets and libraries
import pandas as pd
import numpy as np

df=pd.read_csv('Churn_Modelling.csv')


df.drop(columns=['CustomerId', 'Surname', 'RowNumber'],inplace=True)

# check for duplicates, if exist, remove
df.duplicated().sum()

# clean data, convert to values that algorithms understand
df = pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)

X = df.drop(columns=['Exited'])
y = df['Exited'].values
print(df)

# set test data and training data 
from sklearn.model_selection import train_test_split


# Split data into training, validation, test data. 
# 70% training data, 10% validation data, 20% test data.
X_train_temp, X_test, y_train_temp, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

X_train, X_val, y_train, y_val = train_test_split(X_train_temp, y_train_temp, test_size=0.125, random_state=0)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_trf = scaler.fit_transform(X_train)
X_test_trf = scaler.transform(X_test)

import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(11, activation='sigmoid', input_shape=(11,)))
model.add(Dense(11,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))


model.compile(optimizer='Adamax',loss='binary_crossentropy',metrics=['accuracy'])
# history = model.fit(X_train,y_train,batch_size=50,epochs=100,verbose=1)

input_data = np.array([[ 600, 45, 5, 0.0, 1, 1, 1, 50000.0, False, True, False]])

# Preprocess the input data
input_data = scaler.transform(input_data)  # Assuming 'scaler' is the StandardScaler object you used for scaling your training data

# Make predictions
churn_prediction = model.predict(input_data)

print(churn_prediction)
```
